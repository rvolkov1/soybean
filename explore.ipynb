# Exploratory data analysis

```python
import numpy as np
import polars as pl
```

```python
df = (
  pl.read_csv("nass_data/soybean_yield.csv", separator=",", has_header=True, infer_schema_length=10000, try_parse_dates=True, null_values=["", "(D)", " (D)", "NOT SPECIFIED"])
  .filter(pl.col("Value").is_not_null())
)
df = df[[s.name for s in df if not (s.null_count() == df.height)]]
```
